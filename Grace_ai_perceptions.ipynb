```python
PP = df[['Performance Pressure_1', 'Performance Pressure_2', 'Performance Pressure_3']]
CE = df[['Competency_1', 'Competency_2', 'Competency_3']]
```

```python
import pingouin as pg

pg.cronbach_alpha(PP)
pg.cronbach_alpha(CE)
```

**Cronbach’s alpha for the Performance Pressure (PP) scale was 0.69, indicating marginal internal consistency. While close to the commonly accepted threshold of 0.70, the relatively wide confidence interval suggests variability in item coherence.**

```python
PP12 = df[['Performance Pressure_1', 'Performance Pressure_2']]
CE12 = df[['Competency_1', 'Competency_2']]

pg.cronbach_alpha(PP12)
pg.cronbach_alpha(CE12)
```

**The Performance Pressure scale demonstrated marginal internal consistency (α = 0.69). Item-level analysis indicated that the third item exhibited weaker alignment with the other indicators, suggesting some variation in how different dimensions of performance expectations are perceived.**

DISCUSSION:
- AI may increase different types of pressure, not one single dimension (PP)

**The Competency Escalation scale showed marginal internal consistency, with item-level analysis suggesting that the item related to continuous skill updating may capture a distinct, more long-term dimension of perceived expectations.**

DISCUSSION:
- Students distinguish between needing better skills now vs. constantly updating skills over time (CE)

```python
IV = df[['Perceptions_4', 'Perceptions_5']]
df['IV_score'] = IV.mean(axis=1)
```

```python
df[['IV_score', 'PP_score', 'CE_score']].corr()
```

Initial reliability analysis indicated that the full 5-item scale lacked internal consistency (α = .57). After refinement, a 2-item scale capturing perceived AI integration demonstrated acceptable reliability (α = .73) and was retained for analysis.

Core Finding:
AI is not just increasing one type of expectation, but multiple linked but distinct pressures.

Coorelation Insight:
PP ↔ CE = 0.55 → moderate
IV ↔ PP = 0.46 → solid
IV ↔ CE = 0.28 → weaker

Key pattern:
	•	AI perception → stronger effect on performance pressure (0.46)
	•	AI perception → weaker effect on competency escalation (0.28)

**Perceived AI integration was positively associated with both performance pressure and competency escalation. **While the two constructs were moderately correlated (r = .55), indicating they co-occur, the difference in their relationship with AI perception suggests that they represent distinct dimensions of perceived workplace expectations.**

```python
import statsmodels.api as sm

# Independent variable
X = df['IV_score']
X = sm.add_constant(X)

# Model 1: Performance Pressure
model_PP = sm.OLS(df['PP_score'], X).fit()

# Model 2: Competency Escalation
model_CE = sm.OLS(df['CE_score'], X).fit()

print(model_PP.summary())
print(model_CE.summary())
```

FINDINGS
Perceived AI integration was positively associated with both performance pressure (β = 0.55, p < .001) and competency escalation (β = 0.31, p = .003). However, the relationship was substantially stronger for performance pressure, indicating that students interpret AI primarily as increasing immediate productivity demands rather than long-term skill requirements. The model explained 21.4% of the variance in performance pressure, compared to 7.8% for competency escalation.

⸻

Interpretation

These findings suggest that students perceive AI not only as a driver of increased expectations, but specifically as intensifying short-term performance demands more than long-term skill development. This aligns more closely with work intensification perspectives than with traditional human capital or employability frameworks.
