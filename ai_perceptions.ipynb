# **Cronbach's Alpha**

## Performance Pressure (PP)

Cronbach’s alpha for the Performance Pressure scale indicated good internal consistency **(α = 0.81)**, suggesting that the items reliably capture a common underlying construct related to perceived AI-driven performance expectations.

Item-level analysis showed that the first two items (PP1, PP2) demonstrated strong internal consistency (α = 0.81), while pairwise combinations involving the third item (PP3) were slightly lower (α = 0.68–0.72). This indicates that, although all items contribute to the construct, one item reflects a somewhat distinct aspect of performance pressure.

Overall, the scale can be considered reliable, with minor variation across items reflecting nuanced dimensions of perceived pressure.


## Competency Escalation (CE)

Cronbach’s alpha for the Competency Escalation scale indicated marginal internal consistency **(α = 0.69)**. While close to the commonly accepted threshold of 0.70, this suggests some variability in how consistently the items measure the construct.

Item-level analysis revealed that the first two items (CE1, CE2) showed moderate consistency (α = 0.69), whereas pairwise combinations involving the third item (CE3) were notably lower (α = 0.48–0.59). This suggests that the third item is less aligned with the others.

These findings indicate that the scale may capture related but distinct dimensions of competency escalation, particularly differentiating between immediate skill requirements and longer-term expectations for continuous skill development.

## IV Construct Refinement

The Perceptions scale was originally developed as a broader measure of attitudes toward AI in the workplace for the group project, encompassing multiple dimensions including personal confidence, AI usage in evaluation, and perceived prevalence.

For the purposes of the present analysis, only items directly relevant to the research focus were retained. Specifically, **Perceptions_4 and Perceptions_5 were selected, as they capture perceptions of AI prevalence and integration in the workplace, which aligns with the conceptual definition of the independent variable in this study.**

Items related to personal confidence (Perceptions_1, Perceptions_2) and AI-based evaluation (Perceptions_3) were excluded, as they reflect distinct constructs not central to the current research question.

The resulting **two-item scale demonstrated acceptable internal consistency (α = 0.74)**.

## Regression Analysis

A series of linear regression analyses were conducted to examine the effect of AI Perceptions on Performance Pressure and Competency Escalation.

AI Perceptions significantly predicted Performance Pressure **(β = 0.55, p < .001)**, explaining 22% of the variance **(R² = 0.22)**. This indicates that individuals who perceive AI as more prevalent in the workplace report higher levels of performance-related pressure.

AI Perceptions also significantly predicted Competency Escalation **(β = 0.35, p < .001)**, although the effect was weaker, explaining approximately 10% of the variance **(R² = 0.10)**. This suggests that while perceptions of AI are associated with increased expectations for skill development, the relationship is less pronounced than for performance pressure.

Overall, the findings indicate that AI workplace perceptions are more strongly associated with immediate performance demands than with longer-term competency requirements.

```python
import statsmodels.api as sm

# Independent variable
X = df['IV_score']
X = sm.add_constant(X)

# Model 1: Performance Pressure
model_PP = sm.OLS(df['PP_score'], X).fit()

# Model 2: Competency Escalation
model_CE = sm.OLS(df['CE_score'], X).fit()

# Print results
print("Regression: AI Perceptions & Performance Pressure")
print(model_PP.summary())

print("\n" + "="*80 + "\n")

print("Regression: AI Perceptions & Competency Escalation")
print(model_CE.summary())
```

### Frequency of AI Usage

```python
freq_map = {
    1: 'Never',
    2: 'Sometimes (monthly)',
    3: 'About half the time (weekly)',
    4: 'Most of the time (multiple times a week)',
    5: 'Daily'
}
df['Frequency_label'] = df['Frequency'].map(freq_map)

order = [
    'Never',
    'Sometimes (monthly)',
    'About half the time (weekly)',
    'Most of the time (multiple times a week)',
    'Daily'
]

freq_percent = df['Frequency_label'].value_counts(normalize=True).reindex(order) * 100

colors = plt.cm.Blues(np.linspace(0.15, 0.85, len(freq_percent)))

plt.figure(figsize=(8,8))

plt.pie(
    freq_percent,
    labels=freq_percent.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)

plt.title('Frequency of AI Tool Usage (%)')

# Add annotation for weekly+ usage
weekly_total = freq_percent.loc[
    ['About half the time (weekly)', 'Most of the time (multiple times a week)', 'Daily']
].sum()

plt.text(0, -1.2, f'{weekly_total:.1f}% use AI at least weekly',
         ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

## Analysis with Control Variables

```python
import statsmodels.formula.api as smf
```

```python
# Model 1: Performance Pressure
model_PP = smf.ols(
    formula='PP_score ~ Frequency + Age + Education',
    data=df
).fit()

# Model 2: Competency Expectations
model_CE = smf.ols(
    formula='CE_score ~ Frequency + Age + Education',
    data=df
).fit()

# Print results
print("Multiple Regression: AI Use Frequency & Performance Pressure")
print(model_PP.summary())

print("\n" + "="*80 + "\n")

print("Multiple Regression: AI Use Frequency & Competency Expectations")
print(model_CE.summary())
```

```python
# Model 1: Performance Pressure
model_PP_simple = smf.ols(
    formula='PP_score ~ Frequency',
    data=df
).fit()

# Model 2: Competency Expectations
model_CE_simple = smf.ols(
    formula='CE_score ~ Frequency',
    data=df
).fit()

# Print results
print("Simple Regression: AI Use Frequency & Performance Pressure")
print(model_PP_simple.summary())

print("\n" + "="*80 + "\n")

print("Simple Regression: AI Use Frequency & Competency Expectations")
print(model_CE_simple.summary())
```

Additional analyses treating frequency of AI use as a categorical variable (below) showed a clear increase in performance pressure among respondents who used AI more frequently, particularly at weekly and daily levels.

However, no consistent pattern was observed for competency expectations, suggesting that the relationship between AI use and competency expectations is weak or non-linear.

```python
# Model 1: Performance Pressure
model_PP_simple = smf.ols(
    formula='PP_score ~ C(Frequency)',
    data=df
).fit()

# Model 2: Competency Expectations
model_CE_simple = smf.ols(
    formula='CE_score ~ C(Frequency)',
    data=df
).fit()

# Print results
print("Simple Regression: AI Use Frequency & Performance Pressure")
print(model_PP_simple.summary())

print("\n" + "="*80 + "\n")

print("Simple Regression: AI Use Frequency & Competency Expectations")
print(model_CE_simple.summary())
```

**Performance Pressure & Frequency of AI Use**
- T.3 (weekly): significant
- T.4 (multi-weekly): significant
- T.5 (daily): very significant

Pattern is clear:

The more frequently people use AI, the higher their performance pressure.

Not perfectly linear, but monotonically increasing.

While frequency of AI use was significantly associated with increased perceptions of performance pressure, no consistent relationship was found with competency expectations.

This may be due to the conceptual distinction between the constructs, as competency expectations reflect broader beliefs about employer demands rather than direct personal experiences with AI use.
