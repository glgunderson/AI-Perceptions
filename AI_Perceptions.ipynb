# **Cronbach's Alpha**

## Performance Pressure (PP)

Cronbach’s alpha for the Performance Pressure scale indicated good internal consistency **(α = 0.81)**, suggesting that the items reliably capture a common underlying construct related to perceived AI-driven performance expectations.

Item-level analysis showed that the first two items (PP1, PP2) demonstrated strong internal consistency (α = 0.81), while pairwise combinations involving the third item (PP3) were slightly lower (α = 0.68–0.72). This indicates that, although all items contribute to the construct, one item reflects a somewhat distinct aspect of performance pressure.

Overall, the scale can be considered reliable, with minor variation across items reflecting nuanced dimensions of perceived pressure.


## Competency Expectations (CE)

Cronbach’s alpha for the Competency Expectations scale indicated marginal internal consistency **(α = 0.69)**. While close to the commonly accepted threshold of 0.70, this suggests some variability in how consistently the items measure the construct.

Item-level analysis revealed that the first two items (CE1, CE2) showed moderate consistency (α = 0.69), whereas pairwise combinations involving the third item (CE3) were notably lower (α = 0.48–0.59). This suggests that the third item is less aligned with the others.

These findings indicate that the scale may capture related but distinct dimensions of competency expectations, particularly differentiating between immediate skill requirements and longer-term expectations for continuous skill development.

```python
import statsmodels.api as sm

# Independent variable
X = df['IV_score']
X = sm.add_constant(X)

# Model 1: Performance Pressure
model_PP = sm.OLS(df['PP_score'], X).fit()

# Model 2: Competency Escalation
model_CE = sm.OLS(df['CE_score'], X).fit()

print(model_PP.summary())
print(model_CE.summary())
```

FINDINGS
Perceived AI integration was positively associated with both performance pressure (β = 0.55, p < .001) and competency escalation (β = 0.31, p = .003). However, the relationship was substantially stronger for performance pressure, indicating that students interpret AI primarily as increasing immediate productivity demands rather than long-term skill requirements. The model explained 21.4% of the variance in performance pressure, compared to 7.8% for competency escalation.

⸻

Interpretation

These findings suggest that students perceive AI not only as a driver of increased expectations, but specifically as intensifying short-term performance demands more than long-term skill development. This aligns more closely with work intensification perspectives than with traditional human capital or employability frameworks.
